## Objectron
MediaPipe Objectron is a mobile real-time 3D object detection solution for everyday objects. It detects objects in 2D images, and estimates their poses through a machine learning (ML) model, trained on the Objectron dataset.

## Face Detection
MediaPipe Face Detection is an ultrafast face detection solution that comes with 6 landmarks and multi-face support. It is based on BlazeFace, a lightweight and well-performing face detector tailored for mobile GPU inference. 

```python
from google.colab import files
import cv2
import matplotlib
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
import numpy as np

# ! pip install mediapipe
import mediapipe as mp
mp_face_detection = mp.solutions.face_detection
mp_drawing = mp.solutions.drawing_utils
```

```python
# For Face detection
with mp_face_detection.FaceDetection(
    model_selection=1, min_detection_confidence=0.1) as face_detection:
    image = cv2.cvtColor(cv2.imread('/content/face.jpg'), cv2.COLOR_BGR2RGB)

    # Convert the BGR image to RGB and process it with MediaPipe Face Detection.
    results = face_detection.process(image)
    if not results.detections:
        print("No face detected")
    else: 
        annotated_image = image.copy()
        for detected_object in results.detections:
            mp_drawing.draw_detection(annotated_image, detected_object, 
            mp_drawing.DrawingSpec(color=(121,22,76), thickness=2, circle_radius=4),
            mp_drawing.DrawingSpec(color=(0,255,230), thickness=2, circle_radius=2)
            )
```

```python
plt.figure(figsize=(10,15))
plt.subplot(1,2,1)
plt.imshow(image)

plt.subplot(1,2,2)
plt.imshow(annotated_image)
```
